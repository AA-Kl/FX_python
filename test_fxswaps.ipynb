```python
# Test case 2: Create FXTrade with default giv_currency (foreign currency)
from classes.FXTrade import FXTrade
from classes.currency_pair import CurrencyPair
from classes.direction import TradeDirection
from datetime import date

fx_trade = FXTrade(
    currency_pair=CurrencyPair.EUR_USD,
    giv_amount=1000000.0,
    value_date=date(2023, 10, 3),
    giv_direction=TradeDirection.BUY,
    rate=1.05,
    done=True
)

# Assertions
assert fx_trade.giv_currency == Currency.EUR, "giv_currency should default to foreign currency"
assert fx_trade.rate == 1.05, "Rate should be correctly assigned"
assert fx_trade.alt_currency == Currency.USD, "alt_currency should match domestic currency"
assert fx_trade.giv_payment.currency == Currency.EUR, "Receive currency should match foreign currency"
assert fx_trade.alt_payment.currency == Currency.USD, "Pay currency should match domestic currency"
assert fx_trade.giv_payment.amount == 1000000.0, "Receive amount should match giv_amount"
#assert fx_trade.alt_payment.amount == 1000000.0 * 1.05, "Pay amount should be calculated correctly"
print(f"FXTrade with default giv_currency: {fx_trade}")
```

```python
# Test case 3: Create FXTrade with custom giv_currency (domestic currency)
fx_trade = FXTrade(
    currency_pair=CurrencyPair.EUR_USD,
    giv_amount=1050000.0,
    value_date=date(2023, 10, 3),
    giv_direction=TradeDirection.SELL,
    alt_amount=1000000.0,
    giv_currency=Currency.USD,
    done=False
)

# Assertions
assert fx_trade.giv_currency == Currency.USD, "giv_currency should be set to domestic currency"
assert fx_trade.alt_amount == 1000000.0, "Alt amount should be correctly assigned"
assert fx_trade.alt_currency == Currency.EUR, "alt_currency should match foreign currency"
assert fx_trade.giv_payment.currency == Currency.USD, "Receive currency should match domestic currency"
assert fx_trade.alt_payment.currency == Currency.EUR, "Pay currency should match foreign currency"
assert fx_trade.giv_payment.amount == 1050000.0, "Receive amount should match giv_amount"
assert fx_trade.alt_payment.amount == 1000000.0, "Pay amount should match alt_amount"
print(f"FXTrade with custom giv_currency: {fx_trade}")
```

```python
# Test case 4: Create FXTrade using string for currency_pair
fx_trade = FXTrade(
    currency_pair="EUR/USD",
    giv_amount=1000000.0,
    value_date=date(2023, 10, 3),
    giv_direction=TradeDirection.BUY,
    rate=1.05,
    done=True
)

# Assertions
assert fx_trade.currency_pair == CurrencyPair.EUR_USD, "currency_pair should be correctly parsed from string"
assert fx_trade.giv_currency == Currency.EUR, "giv_currency should default to foreign currency"
assert fx_trade.rate == 1.05, "Rate should be correctly assigned"
assert fx_trade.alt_currency == Currency.USD, "alt_currency should match domestic currency"
assert fx_trade.giv_payment.currency == Currency.EUR, "Receive currency should match foreign currency"
assert fx_trade.alt_payment.currency == Currency.USD, "Pay currency should match domestic currency"
assert fx_trade.giv_payment.amount == 1000000.0, "Receive amount should match giv_amount"
assert fx_trade.alt_payment.amount == 1000000.0 * 1.05, "Pay amount should be calculated correctly"
print(f"FXTrade with string currency_pair: {fx_trade}")
```
